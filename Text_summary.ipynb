```python
import csv
from llmx import llm, TextGenerationConfig
import os

csv_file_path = "/kaggle/input/covid-19/Covid-19.csv"

csv_data = []
with open(csv_file_path, newline="", encoding="utf-8") as file:
    reader = csv.reader(file)
    for row in reader:
        csv_data.append(row)

system_message = {"role": "system", "content": "You are an experienced data analyst that can perform analysis for states of USA and the values associated in the CSV file. Give me conclusions from the data you receive and also perform geographic analysis with the use of this data."}

user_messages = []
for row in csv_data:
    content = ", ".join(row)
    user_messages.append({"role": "user", "content": content})

messages = [system_message] + user_messages
```

```python
messages
```

```python
config = TextGenerationConfig( 
    n=1,
    temperature=0.8,
    max_tokens=100,
    top_p=1.0,
    top_k=50,
    frequency_penalty=0.0,
    presence_penalty=0.0,
)

hf_generator = llm(provider="hf", model="HuggingFaceH4/zephyr-7b-beta", device_map="auto")
hf_config = TextGenerationConfig(temperature=0, max_tokens=650, use_cache=False)
hf_response = hf_generator.generate(messages, config=hf_config)
print(hf_response.text[0].content)
```
